```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#read the data from csv and convert the data to dataframe
game=pd.read_csv('question2_data.csv',header=None)

#add the column names to dataframe
game.columns=['id','win','loss','tie']

#reindex the dataframe's index from 1
#easier to check the number of game
game.index=np.arange(1,len(game)+1)

#show game dataframe
game
```

```python
#convert each columns to the lists
team_list=game['id'].tolist()
win_list=game['win'].tolist()
loss_list=game['loss'].tolist()
tie_list=game['tie'].tolist()

#initial values for the variables
total_game=0
win_avg=0
wip_total=0

#use a for loop to process each team in the team list
for i in range(len(team_list)):
    
    #total game = wins+losses+ties
    #since each list should be same length and the [i] should be match between team list and game lists
    #for example: team_list[0] is 9867, in the dataframe the win=4, so the win_list[0] is 4
    #same for loss_list and tie_list
    
    total_game=win_list[i]+loss_list[i]+tie_list[i]
    
    #if total game is not euqal to 0, the program will calculate game results for each team
    if total_game!=0:
        
        #print team information first
        
        print(f"Team: {team_list[i]}\n{win_list[i]} wins    {loss_list[i]} losses     {tie_list[i]} ties")
        
        #If the total number of games played is exactly 16, the program will print a message saying the team’s season is finished.
        if total_game==16:
            print(f"Total number of games played is {total_game}    The team’s season is finished")
    
        #If the total is less than 16, the program will printhow many games are still remaining in the season.
        if total_game<16:
            print(f"Total number of games played is {total_game}    {16-total_game} games still remaining")
        
        #The winning average is defined to be the number of games won divided by the total number of games played.
        win_avg=win_list[i]/total_game
        
        #The program should print this out as shown here,with exactly 4 decimal places.
        print(f"The winning average is {win_avg:.4f}")
        
        
        #The program will compare the number of games tied to both the number of games won and the number of games lost
        
        #the program will determine whether or not the number of games tied is greater than or equal to the number won. 
        
        if tie_list[i]>win_list[i]:
            print('Number of games tied is greater than number won')
        elif tie_list[i]==win_list[i]:
            print('Number of games tied is equal to number won')
        elif tie_list[i]<win_list[i]:
            print('Number of games tied is not greater than or equal to number won')
            
        #Then the program will determine whether or not the number of games tied is greater than the number lost. 
        
        if tie_list[i]>loss_list[i]:
            print('Number of games tied is greater than number lost')
        elif tie_list[i]<=loss_list[i]:
            print('Number of games tied is not greater than number lost')
    
        #wip total is the number of wins plus the number ties minus three times the number of losses
        wip_total=(win_list[i]+tie_list[i])-3*loss_list[i]
        
        #the value cannot be negative. If it would evaluate to a negative value, the program should display it as 0.
        if wip_total<0:
            wip_total=0
        
        #display wip total message
        print("The wip total is ",wip_total)
    
        #new line between each team
        print('\n')
        
        #reset the initial values to 0 for variables then the program can calculate the game results for new team correctly
        
        total_game=0
        win_avg=0
        wip_total=0
    
    #if total game is euqal to 0, the program will display messages directly
    else:
        print(f"Team: {team_list[i]}\n{win_list[i]} wins    {loss_list[i]} losses     {tie_list[i]} ties")
        print(f"Total number of games played is {total_game}    {16-total_game} games still remaining")
        print(f"The winning average is {win_avg:.4f}")
        print(f"No comparative records of ties and wins")
        print(f"No comparative records of ties and losses")
        print(f"No records of wip total")
        print('\n')

#after all teams processed in the team list, the for loop will stop
#then the all teams processed and all information displayed
print('Last team has been processe')

#print the total number of teams that have been processed, then stop
#since the dataframe's column convert to list, so the length of team list should be the number of teams
team_num=len(team_list)
print("Total team numbers: ",team_num)
print("This program stop")
```
